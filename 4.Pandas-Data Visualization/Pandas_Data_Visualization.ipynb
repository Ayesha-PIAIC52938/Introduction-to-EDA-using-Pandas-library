###  <div class = "alert alert-info"> groupby() function (aka. grouping in pandas)
The<span style="color:red"> groupby()</span> method allows you to group your data and execute functions on these groups.

<b><span style="color:green"> Syntax:</span></b>
    <b> dataframe.groupby(by, axis, level, as_index, sort, group_keys, observed, dropna)</b>
    
    
<b>Parameters:</b>
    
The<span style="color:red"> axis, level, as_index, sort, group_keys, observed, dropna</span> parameters are keyword arguments.
    
|Parameter|Value|Description|    
|---------|-----|-----------|
|by       |     |Required. A label (row or column), a list of labels, or a function used to specify how to group the DataFrame.|
|axis	  |  0,1,'index', 'columns'|Optional, Which axis to make the group by, default 0. |
|level|level,None|Optional. Specify if grouping should be done by a certain level. Default None|
|as_index|True,False|Optional, default True. Set to False if the result should NOT sort the group keys (for better performance)|
|group_keys|True,False|Optional, default True. Set to False if the result should NOT add the group keys to index|
|dropna|True,False|Optional, default True. Set to False if the result should include the rows/columns where the group key is a NULL value|

<b>Return value:</b>
A DataFrameGroupBy object where the rows/columns are grouped.

#### <div class = "alert alert-warning"> Ride count per year for four years Plot :<div >

```python
trip_count_data_per_year = data_frame.groupby(['year'])["trip_id"].count()   #trip_count_data is a series

#(['year'])------------------->groupby is done on year column
#["trip_id"].count() --------->counting nomber of trips against each yeat in the year col

trip_count_data_per_year     #bicycle rides count per year
```

```python
trip_count_data_per_year.index
```

```python
plt.plot(trip_count_data_per_year.index,trip_count_data_per_year,label='ride count per year')
# naming the x axis
plt.xlabel('years')
# naming the y axis
plt.ylabel('rides')
# giving a title to my graph
plt.title('ride count per year')
  
# show a legend on the plot
plt.legend()
plt.show
```

### <div class = "alert alert-warning"> Ride count per month for four years Plot :<div >

```python
trip_count_data_per_month = data_frame.groupby(['month'])["trip_id"].count()   # bicycle rides count per month (of 4 years)
trip_count_data_per_month     
```

```python
plt.plot(trip_count_data_per_month.index,trip_count_data_per_month,label='ride count per month')

plt.xlabel('years')# naming the x axis
plt.ylabel('rides')# naming the y axis
plt.title('ride count per month')# giving a title to my graph
plt.legend()# show a legend on the plot

plt.show
```

### <div class = "alert alert-warning"> Ride count per month for the years 2014 Plot :<div >

### <div class = "alert alert-warning"> Ride count in January for the years 2014 Plot :<div >

```python
trip_count_data_for_jan_2014 = data_frame[(data_frame['year']==2014) & (data_frame['month']== 1)].groupby(['day'])["trip_id"].count()

trip_count_data_for_jan_2014  # assuming monday being first day of the week i.e 0 : Monday ---- 6 : Sunday
```

```python
plt.plot(trip_count_data_for_jan_2014.index,trip_count_data_for_jan_2014,label='ride count for January of 2014')

plt.xlabel('years')# naming the x axis
plt.ylabel('rides')# naming the y axis
plt.title('ride count i january for the year 2014')# giving a title to my graph
plt.legend()# show a legend on the plot

plt.show
```

### <div class = "alert alert-warning"> Ride count in January, Faburary and March for the years 2014 Plot :<div >

```python
trip_count_data_for_jan_2014 = data_frame[(data_frame['year']==2014) & (data_frame['month']== 1)].groupby(['day'])["trip_id"].count()
trip_count_data_for_fab_2014 = data_frame[(data_frame['year']==2014) & (data_frame['month']== 2)].groupby(['day'])["trip_id"].count()
trip_count_data_for_mar_2014 = data_frame[(data_frame['year']==2014) & (data_frame['month']== 3)].groupby(['day'])["trip_id"].count()
```

```python
plt.plot(trip_count_data_for_jan_2014.index,trip_count_data_for_jan_2014,label='ride count in January of 2014')
plt.plot(trip_count_data_for_fab_2014.index,trip_count_data_for_fab_2014,label='ride count in Faburary of 2014')
plt.plot(trip_count_data_for_mar_2014.index,trip_count_data_for_mar_2014,label='ride count in March of 2014')


plt.xlabel('years')   # naming the x axis
plt.ylabel('rides')  # naming the y axis
plt.title('Ride count for January-March in the year 2014')# giving a title to my graph
plt.legend()# show a legend on the plot

plt.show
```

### <div class = "alert alert-success"> Bubble Plot :
for 3 dimension plotting

```python
# Import the required modules
import matplotlib.pyplot as plt
import numpy as np

# Set figure size
plt.figure(figsize=(8,5))

# Create the data
countries = ['Qatar','Luxembourg','Singapore','Brunei','Ireland','Norway','UAE','Kuwait']
populations = [2781682, 604245,5757499,428963,4818690,5337962,9630959,4137312]
gdp_per_capita = [130475, 106705, 100345, 79530, 78785, 74356,69382, 67000]

# scale GDP per capti income to shot the bubbles in the graph
scaled_gdp_per_capita  = np.divide(gdp_per_capita, 80)

colors = np.random.rand(8)

# Draw the scatter diagram
plt.scatter(countries, populations, s=scaled_gdp_per_capita, c=colors, cmap="Blues",edgecolors="grey", alpha=0.5)

# Add X Label on X-axis
plt.xlabel("Countries")

# Add Y Label on X-axis
plt.ylabel("Population")

# Add title to graph
plt.title("Bubble Chart")

# rotate x label for clear visualization
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
